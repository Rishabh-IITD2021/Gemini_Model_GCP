```python
# Assuming 'df' is already defined and loaded as in the previous response.
value_counts = df['Workspace name'].value_counts() # Replace 'Column_Name' with the actual column you want to analyze
value_counts
```

```python
# Filter out rows where 'Epic Name' is empty
df_filtered = df[df['Epic name'].notna()]
```

```python
len(df_filtered)
```

```python
# Combine 'Feature Name' and 'Requirement Name' to create a new column
df_filtered['Combined Name'] = df_filtered['Feature name'].combine_first(df['Requirement name'])

# Filter out rows where both 'Feature Name' and 'Requirement Name' are NaN
df_filtered = df_filtered.dropna(subset=['Combined Name'])

# Optionally drop the combined column if you don't need it anymore
df_filtered = df_filtered.drop(columns=['Combined Name'])

```

```python
# Now df_filtered will contain rows where at least one of 'Feature Name' or 'Requirement Name' has a value
df_filtered
```

```python


# Combine 'Feature Name' and 'Requirement Name' to create a new column
df_filtered['Combined Name'] = df_filtered['Feature description'].combine_first(df['Requirement description'])

# Remove rows where 'Combined Name' is NaN
df_filtered = df_filtered.dropna(subset=['Combined Name'])

# Optionally drop the combined column if you don't need it anymore
df_filtered = df_filtered.drop(columns=['Combined Name'])
```

```python
df_filtered
```

```python

# Create the new DataFrame
new_df = pd.DataFrame()

# Add other columns
new_df['Workspace Name'] = df_filtered['Workspace name']
new_df['Epic Name'] = df_filtered['Epic name']

# Combine 'Feature Name' and 'Requirement Name'
new_df['Objective '] = df_filtered['Feature name'].fillna('') + ' \n ' + df_filtered['Requirement name'].fillna('')

# Combine 'Feature Description' and 'Requirement Description'
new_df['Final Requirement Description'] = df_filtered['Feature description'].fillna('') + ' \n' + df_filtered['Requirement description'].fillna('')


#Remove rows where combined name is empty
new_df = new_df[new_df['Final Requirement Description'] != ' - ']


# Display the new DataFrame
new_df.reset_index(drop=True, inplace=True)
```

```python
new_df
```

```python
# Group by 'Workspace name'
grouped = new_df.groupby('Workspace Name')

sampled_data = []

# Try to get 10 samples from each workspace
for workspace, group_data in grouped:
  if len(group_data) >= 20:
      sampled_data.extend(group_data.sample(20).to_dict('records'))
  elif len(group_data) > 0:  #Take all if less than 20
      sampled_data.extend(group_data.to_dict('records'))

# Check if we have at least 100 samples; if not, take more samples randomly
if len(sampled_data) < 100:
    remaining_samples = 100 - len(sampled_data)
    available_samples = df_filtered[~df_filtered.index.isin([x['Unnamed: 0'] for x in sampled_data])]
    if len(available_samples) > 0: # Check if any available samples
        sampled_data.extend(available_samples.sample(min(remaining_samples, len(available_samples))).to_dict('records'))

sampled_df_1 = pd.DataFrame(sampled_data)

```
